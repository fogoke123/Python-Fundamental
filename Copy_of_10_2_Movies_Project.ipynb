### Plot a bar chart of mean rank and genre
---

```python
from pandas._libs.algos import rank_1d
from scipy.stats import linregress
from scipy.stats.morestats import Mean
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
movies_data = pd.read_csv("https://github.com/lilaceri/Working-with-data-/blob/main/Data%20Sets%20for%20code%20divisio/movies.csv?raw=true")
def mean_rank_genre():
  mean_rank_genre = movies_data.groupby('genre').mean()
  mean_rank_genre_df = mean_rank_genre['rank']
  label = mean_rank_genre_df.keys()
  plt.bar(label, mean_rank_genre_df)
  plt.xlabel("rank")
  plt.ylabel("genre")
  plt.xticks(rotation=90)
  plt.show()
  


mean_rank_genre()
```
